```python
from selenium.webdriver import Chrome

driver = Chrome()
driver.get('https://www.holidify.com/places/amritsar/sightseeing-and-things-to-do.html')
```

```python
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time
wait=WebDriverWait(driver,10) # explicit wait
```

```python
data=[]
filter=driver.find_element(By.CLASS_NAME,'category-filter')
filter.click()
time.sleep(1)
tags = driver.find_elements(By.NAME,'ptvfilterTag')
checkboxes = driver.find_elements(By.CLASS_NAME,'checkbox-hole')
values=[]
for i in range(len(tags)):
    value = tags[i].get_attribute("value")
    values.append(value)
    checkboxes[i].click()
    time.sleep(1)
    apply = driver.find_element(By.ID, 'applyModalFilter')
    apply.click()
    time.sleep(2)
    places = driver.find_elements(By.CLASS_NAME, "card-heading")
    for rank in range(1,len(places)+1):
        row=[]
        row.append(places[rank-1].text)
        row.append(value)
        row.append(rank)
        data.append(row)
        print(row)
    #scrape
    #uncheck
    filter.click()
    time.sleep(1)
    checkboxes[i].click()
```

```python
import pandas as pd
import copy
d={"Shopping": "shopping", "Gurudwara" : "religious", "Temple": "religious", "Mosque" : "religious",  "Amusement & Theme Park" : "entertainment","Historical Site" : "historical site","Forts & Palaces" : "historical site","Museum" : "historical site", "Garden & Park" : "park", "Bird Sanctuary" : "park", "Shows & Theatres" : "entertainment"}
df = pd.DataFrame(data)
data_copy = copy.deepcopy(data)
for row in data_copy : 
    val = row[1]
    print(val)
    if val != "Select":
        row[1] = d[val]
df1=pd.DataFrame(data_copy, columns=['place','category','rank'])
df1
```

```python
religious = df1[df1['category'] == 'religious']
historic = df1[df1['category'] == 'historical site']
entertainment = df1[df1['category'] == 'entertainment']
shopping = df1[df1['category'] == 'shopping']
park = df1[df1['category'] == 'park']
```

```python
df.to_csv('master.csv')
religious.to_csv('religious.csv')
historic.to_csv('historic.csv')
entertainment.to_csv('entertainment.csv')
shopping.to_csv('shopping.csv')
park.to_csv('park.csv')
```
